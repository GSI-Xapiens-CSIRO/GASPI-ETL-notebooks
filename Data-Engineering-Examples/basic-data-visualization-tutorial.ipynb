## Basic Data Visualization: Line Charts and Bar Graphs with Matplotlib


## In this tutorial, we will learn how to:
#### - Create simple line charts
#### - Create bar graphs
 
## Let's get started by importing the Matplotlib library.

### 1. Importing Matplotlib
#### Matplotlib is a powerful library for creating visualizations in Python.

```python
import matplotlib.pyplot as plt
```

### 2. Creating a Line Chart
#### Line charts are great for visualizing trends over time or continuous data.

```python
# Example data
years = [2018, 2019, 2020, 2021, 2022]
sales = [250, 300, 400, 450, 500]

# Create a line chart
plt.figure(figsize=(8, 5))  # Set the figure size
plt.plot(years, sales, marker='o', linestyle='-', color='b', label="Sales")

# Add labels and title
plt.title("Yearly Sales Over Time", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Sales (in $1000s)", fontsize=12)

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

### 3. Creating a Bar Graph
#### Bar graphs are useful for comparing categories of data.

```python
# Example data
categories = ["Electronics", "Clothing", "Furniture", "Toys"]
profits = [50000, 35000, 40000, 20000]

# Create a bar graph
plt.figure(figsize=(8, 5))  # Set the figure size
plt.bar(categories, profits, color='orange', alpha=0.8)

# Add labels and title
plt.title("Profit by Product Category", fontsize=14)
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Profit (in $)", fontsize=12)

# Show the plot
plt.show()
```

### 4. Adding Customizations
#### You can customize your visualizations to make them more informative.

```python
# Example: Adding custom colors and labels to a bar graph
colors = ['red', 'blue', 'green', 'purple']
plt.figure(figsize=(8, 5))
plt.bar(categories, profits, color=colors, alpha=0.9)

# Add value labels above bars
for i, value in enumerate(profits):
    plt.text(i, value + 1000, f"${value}", ha='center', fontsize=10)

# Add labels and title
plt.title("Profit by Product Category with Value Labels", fontsize=14)
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Profit (in $)", fontsize=12)

# Show the plot
plt.show()
```
